```python
#import libraries
import pandas as pd
import numpy as np
```

```python
# Load the dataset
anime_df = pd.read_csv("C:\\Users\\dheni\\Downloads\\DS Assignments\\11.Recommendation System\\anime.csv")
anime_df
```

```python
#missing values
anime_df.isnull().sum()
```

```python
# Handle missing values
anime_df.fillna('', inplace=True)
```

```python
# Explore the dataset
anime_df.info()
```

```python
#import library
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Combine genres and ratings into a single feature
anime_df['features'] = anime_df['genre'] + ' ' + anime_df['rating'].astype(str)
anime_df
```

```python
# Convert genres to numerical representations using TF-IDF
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(anime_df['features'])
tfidf_matrix.shape
```

```python
#import library
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Compute the cosine similarity matrix
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
cosine_sim
```

```python
# Function to get recommendations
def get_recommendations(name, cosine_sim=cosine_sim):
    # Get the index of the anime that matches the title
    idx = anime_df[anime_df['name'] == name].index[0]
    # Get the pairwise similarity scores of all animes with that anime
    sim_scores = list(enumerate(cosine_sim[idx]))
    # Sort the animes based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    # Get the scores of the 10 most similar animes
    sim_scores = sim_scores[1:11]
    # Get the anime indices
    anime_indices = [i[0] for i in sim_scores]
    # Return the top 10 most similar animes
    return anime_df['name'].iloc[anime_indices]
# Test the recommendation system
print(get_recommendations('Naruto'))
```

```python
#import libraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, f1_score
import warnings
warnings.filterwarnings('ignore')
```

```python
# Split the dataset into training and testing sets
train, test = train_test_split(anime_df, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
# Prepare lists for true and predicted values
y_true = []
y_pred = []
mlb = MultiLabelBinarizer()
```

```python
# Specify the target names in the test set
for name in test['name']:
    # Get the index of the name in the anime DataFrame
    target_index = anime_df[anime_df['name'] == name].index[0]
    # Get similarity scores for the target anime
    sim_scores = list(enumerate(cosine_sim[target_index]))
    # Sort the anime based on similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    # Get the top 5 similar anime (excluding the target itself)
    top_n = 5
    sim_scores = sim_scores[1:top_n + 1]  # Exclude the target anime itself
    anime_indices = [i[0] for i in sim_scores]
    # Append the actual name for evaluation
    y_true.append(name)  # Store the true name
    # Store the predicted names
    predicted_names = anime_df.iloc[anime_indices]['name'].tolist()
    y_pred.append(predicted_names)
```

```python
# Transform y_true and y_pred to a binary format
y_true_binary = mlb.fit_transform(y_true)
y_pred_binary = mlb.transform([names for names in y_pred])
```

```python
#Evaluating precision, recall, and F1-score
precision = precision_score(y_true_binary, y_pred_binary, average='micro')
recall = recall_score(y_true_binary, y_pred_binary, average='micro')
f1 = f1_score(y_true_binary, y_pred_binary, average='micro')
```

Can you explain the difference between user-based and item-based collaborative filtering?

User-based collaborative filtering recommends items to a user based on the preferences of similar users. It looks for users who have a similar taste and suggests items that those similar users liked.

Item-based collaborative filtering recommends items based on the similarity of items themselves. It uses the rating patterns of users across items to find items that are similar and recommends them.

What is collaborative filtering, and how does it work?

Collaborative filtering is a method used by recommender systems to predict a user's interests by collecting preferences from many users. It assumes that if two users agree on one issue, they are likely to agree on others as well. This can be done using either user-based or item-based methods.
